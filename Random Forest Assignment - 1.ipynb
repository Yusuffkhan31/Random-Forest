```python
bins = [-1,12,18]
Sales_cat = ['Low','High']
data['Sales_categorical'] = pd.cut(data['Sales'], bins, labels = Sales_cat)
```

```python
data.head()
```

```python
column = data.pop('Sales_categorical')
```

```python
data.insert(0, 'Sales_categorical', column)
```

```python
data1 = data.drop('Sales', axis=1)
data1.head()
```

```python
#Label Encoding
label_encoder = preprocessing.LabelEncoder()
data1['Sales_categorical']= label_encoder.fit_transform(data1['Sales_categorical'])
```

```python
data_encoded = pd.get_dummies(data1)
data_encoded.head()
```

```python
array = data_encoded.values
```

```python
X = array[:,1:8]
y = array[:,0]
```

```python
#Random Forest
seed = 7
```

```python
kfold = KFold(n_splits=10,shuffle=True,random_state=seed)
kfold
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
num_trees = 100
max_features = 3
```

```python
model_random = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
```

```python
results = cross_val_score(model_random,X,y,cv=kfold)
print(results.mean())
```
